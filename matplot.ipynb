```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[10,20,25,30,40]

plt.plot(x,y)

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Simple Line Plot")

plt.show
```

```python
import numpy as np

x = np.arange(-10,11,1)
y = x ** 2

plt.plot(x,y,marker="o",linestyle="-",color= "b",label = "y = x^2")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Plot of y =  x²")
plt.legend()
plt.grid(True)

plt.show
```

```python
# Q3: Generate a scatter plot with 20 random points (0, 50) for both x and y.

import numpy as np
x = np.random.randint(0,50,20)
y = np.random.randint(0,50,20)

plt.scatter(x , y , color = "blue",marker="o")
plt.xlabel("X value")
plt.ylabel("Y value")
plt.title("Random Scatter Plot")
plt.grid(True)
plt.show()
```

```python
# Q4: Plot a pie chart showing fruit distribution (Apple: 30%, Banana: 25%, Mango: 20%, Grapes: 15%, Orange: 10%).

import matplotlib.pyplot

data = [30,25,20,15,10]
labels = ["Apple" , "Banana" , "Mango" , "Grapes" , "Orange"]
colors= ["red" , "yellow" , "blue","green" , "orange"]
plt.pie(data, labels=labels, autopct='%1.1f%%', colors=colors, startangle=140)
plt.title("Fruit Distribution")
```

```python
#  Q5: create a histrogram for a dataset of 1000 random number (1 to 1000)
import numpy as np

data = np.random.randint(1,101,1000)
plt.hist(data , bins=20, color="red", edgecolor = "black")
plt.xlabel("Range")
plt.ylabel("Frequency")
plt.title("Histogram of Random Data")

```

```python
# Q5: Create a filled area plot for y = sin(x) 
x=np.linspace(0,10,100)
y= np.sin(x)

plt.fill_between(x,y,color = "skyblue",alpha = 0.5)
plt.plot(x,y,color = "blue",linewidth = 2)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.title(" Filled Area Plot of sin(x) ")
```
